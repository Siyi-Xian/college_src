## Problem 5

Import wine data from website using numpy and process data like in project 1. The only chaging parameter is $Gamma$. I choose to start from $0.02$ to $20$ and step will be $0.02$.

```
import numpy as np

# import and load data
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'
data = np.loadtxt(url, delimiter=",")

# divide data into X and y
X = data[:, 1:14]
y = data[:, 0].astype(int)
# scale all X
for i in range(13):
  X[:, i] = X[:, i] / max(X[:, i])
  
from sklearn.svm import SVC
from sklearn.model_selection import cross_validate

ran = range(1000)
step = 0.02
scores_list = list(ran)
max_index = 0
for i in ran:
  svm = SVC(kernel='rbf', random_state=0, gamma=(i + 1) * step, C=1.0)
  svm_scores = cross_validate(svm, X, y, cv=5, scoring='accuracy')
  scores_list[i] = svm_scores['test_score'].mean()
  if svm_scores['test_score'].mean() > scores_list[max_index]:
    max_index = i

print("Mean:", np.mean(scores_list))
print("Standard Deviation:", np.std(scores_list))
print("Maxima Difference:", max(scores_list) - min(scores_list))
print("Maxima Gama:", (max_index + 1) * step, "with accuracy:", scores_list[max_index])
  
plt.xlabel('Gamma')
plt.ylabel('Accuracy')
plt.title('Learning Curves (SVM, Gamma)')
  
plt.plot([(i + 1) * step for i in ran], scores_list)
plt.show()
```

As graph plot above we know that the maxima accuracy is $1.32$. And apply to the modle the accuracy will be $98.3$.

```
svm = SVC(kernel='rbf', random_state=0, gamma=1.32, C=1.0)
svm_scores = cross_validate(svm, X, y, cv=5, scoring='accuracy')

from sklearn.neural_network import MLPClassifier
# initializate classifier 
mlp = MLPClassifier(solver = 'lbfgs', 
                    activation = 'logistic', 
                    hidden_layer_sizes = (52), 
                    learning_rate_init = 1e-5, 
                    max_iter = 1e+4,
                    verbose = True)
mlp_scores = cross_validate(mlp, X, y, cv=5, scoring='accuracy')

import pandas as pd

columns = mlp_scores.keys()
index = ['MLP', 'SVM']
result_mtx = pd.DataFrame([[np.mean(score) for score in mlp_scores.values()],
                           [np.mean(score) for score in svm_scores.values()]],
                          columns = columns, index = index)
print(result_mtx)
```

Thus, we can find MLP is much slower than SVM and even has lower accuracy than SVM.
